```python
plt.figure(figsize=(20,20))
# sns.scatterplot(df.Store_Area, df.Store_Sales ,alpha=0.6)
sns.scatterplot(df.Store_Area, df.Items_Available,hue=df.Store_Sales, palette='rainbow' ,alpha=0.6)
plt.title("Difference between charges of smokers and charges of non-smokers is apparent")
plt.show()
```

```python
# this plot show as the store area increase the available items icrease as well and that is normal to fill the store capacity 


# regarding the store sales doesn't seem to change much depending on these

# we can tell that the grdient on the right slightly showes more sales but it's not a major thing to consider 
```

```python
plt.figure(figsize=(25,20))

plt.subplot(4,1,1)
a = df.groupby('Store_Area')['Daily_Customer_Count'].mean()
plt.plot(a, color='black')
plt.title('Average customer count per soter area Trend')
plt.xlabel('Store_Area')
plt.ylabel('Daily_Customer_Count')



plt.subplot(4,1,2)
a = df.groupby('Store_Area')['Store_Sales'].mean()
plt.plot(a, color='black')
plt.title('Average store salesper Area  Trend')
plt.xlabel('Store_Area')
plt.ylabel('Store_Sales')



plt.subplot(4,1,3)
a = df.groupby('Store_Area')['Items_Available'].mean()
plt.plot(a, color='black')
plt.title('Average Items Available  Trend')
plt.xlabel('Store_Area')
plt.ylabel('Items_Available')




plt.subplot(4,1,4)
a = df.groupby('Items_Available')['Store_Sales'].mean()
plt.plot(a, color='black')
plt.title('Average Items Available to sales  Trend')
plt.xlabel('Items_Available')
plt.ylabel('Store_Sales')
```

```python
# from these trends we can get that increasing the store size is not affecting the sales or the number of customers in the store 
```
